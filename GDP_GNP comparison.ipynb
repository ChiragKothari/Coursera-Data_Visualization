# Comparison between GDP, GNP, NNP, NI, PI.
 The data is taken from https://www.bea.gov/national/index.htm#gdp, - Current-dollar and "real" GDP.
 GDP - Gross Domestic Product,
 GNP - Gross National Product,
 NNP - Net National Product,
 NI - National Income,
 PI - Personal Income.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
gdp = pd.read_excel(r"C:\Users\Administrator\Downloads\gdplev.xlsx", skiprows = 7)
gdp.head()
```

```python
gdp = gdp[['Unnamed: 0', 'Unnamed: 1']]
gdp.rename(columns = {'Unnamed: 0' : 'Year', 'Unnamed: 1' : 'GDP'}, inplace = True)
gdp = gdp.astype(float)
gdp.set_index('Year', inplace = True)
gdp = gdp[gdp.index > 1987]
gdp.head()
```

```python
df = pd.read_excel("C:\\Users\\Administrator\\Downloads\\Section1All_xls.xlsx", 'T10705-A', skiprows = 7).copy()
df.head()
```

```python
df = df.drop(columns = ['Line', 'Unnamed: 2'])
df.set_index('Unnamed: 1', inplace = True)
df = df.transpose()
df = df.reset_index()
df.head()
```

```python
df.columns
```

```python
df = df[['index', 'Equals: Gross national product', 'Equals: Net national product', 'Equals: National income',
         'Equals: Personal income']]
df.rename(columns = {'index' : 'Year', 'Equals: Gross national product' : 'GNP', 'Equals: Net national product' : 'NNP',
         'Equals: National income' : 'NI', 'Equals: Personal income' : 'PI'}, inplace = True)
df = df.astype(float)
df = df.set_index('Year')
df = df/1000
df = df[df.index > 1987]
df.head()
```

```python
data = gdp.join(df)
data.head()
```

```python
x = pd.tools.plotting.scatter_matrix(data)
```
