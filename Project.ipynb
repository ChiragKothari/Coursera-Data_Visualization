# Comparison between various factors related to GDP
## Percentage change from preceding period in prices for Gross Domestic Purchases.
The data is taken from https://www.bea.gov/national/index.htm#gdp - Current-dollar and "real" GDP.
Comparison between
Motor vehicles and parts,
Gasoline and other energy goods,
Household consumption expenditures,
Health care,
Fixed investment.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
df = pd.read_excel(r"C:\Users\Administrator\Downloads\Section1All_xls.xlsx", 'T10608-A', skiprows = 7, index_col = 1)
df.head()
```

```python
df = df.drop(columns = ['Line', 'Unnamed: 2'])
df = df.transpose()
df.index.name = 'Year'
df.head()
```

```python
df = df[['      Motor vehicles and parts', '      Gasoline and other energy goods',
         '    Household consumption expenditures','      Health care', '  Fixed investment']]
df = df.rename(columns = {'      Motor vehicles and parts' : 'Vehicles',
                          '      Gasoline and other energy goods' : 'Energy Goods',
                          '    Household consumption expenditures' : 'Household (con. & Exp.)',
                          '      Health care' : 'Health Care', '  Fixed investment' : 'Fixed Investment'})
df = df.astype(float)
df.head()
```

```python
pd.tools.plotting.scatter_matrix(df)
```

```python
import seaborn as sns
g = sns.pairplot(df, diag_kind='kde', size=2, plot_kws = {'alpha' : 0.5});
```
